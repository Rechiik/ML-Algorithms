## üìà Simple Linear Regression

Simple Linear Regression is a statistical method used to model the relationship between a single input variable x and a continuous output variable y using a straight line.

---

### üßÆ Hypothesis Function

$$
\hat{y} = \theta_0 + \theta_1 x
$$

---

### üí∏ Cost Function (Mean Squared Error)

$$
J(\theta_0, \theta_1) = \frac{1}{2m} \sum_{i=1}^{m} \left( \hat{y}^{(i)} - y^{(i)} \right)^2
$$


### üîÅ Gradient Descent

An iterative algorithm to minimize the cost function by updating the parameters:

### üîß Parameter Update Rules:

$$
\theta_0 := \theta_0 - \alpha \cdot \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)})
$$

$$
\theta_1 := \theta_1 - \alpha \cdot \frac{1}{m} \sum_{i=1}^{m} (\hat{y}^{(i)} - y^{(i)}) \cdot x^{(i)}
$$

Where:
- $$ \alpha :  learning  rate $$

---

### Algorithm

1. **Initialize** parameters $ \theta_0 $, $ \theta_1 $ to zero or small random values.
2. **Repeat** for a number of iterations (or until convergence):
   - Compute predictions: $ \hat{y}^{(i)} = \theta_0 + \theta_1 x^{(i)} $
   - Compute gradients:
     - $$ \frac{\partial J}{\partial \theta_0} $$
     - $$ \frac{\partial J}{\partial \theta_1} $$
   - Update parameters using the formulas above
3. **Output** the learned values of $ \theta_0 $, $ \theta_1 $


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, mean_absolute_percentage_error
```

```python
x = np.array([0, 1, 2, 3, 4, 5]).reshape(-1, 1)  # Features must be 2D
y = np.array([4, 6, 8, 10, 12, 14])
```

```python
model = LinearRegression()
model.fit(x, y)
```

```python
l = np.array([5,7]).reshape(-1, 1)
y_pred = model.predict(l)
y_pred
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', label='Predicted Line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y = [20, 26, 30, 16, 22, 8]
x = np.array([8, 11, 13, 6, 9, 2]).reshape(-1, 1)
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)
mape = mean_absolute_percentage_error(y, y_pred)

# Print metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2%}")
print(f"R¬≤ Score: {r2:.2f}")
```
